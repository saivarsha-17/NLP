```python
import spacy
import pytextrank
```

```python
document = """Not only did it only confirm that the film would be unfunny
and generic, but it also managed to give away the ENTIRE movie;
and I'm not exaggerating - every moment, every
plot point,every joke id told in the trailer."""
```

```python
en_nlp= spacy.load("en_core_web_sm")
en_nlp.add_pipe("textrank")
doc=en_nlp(document)
```

```python
tr = doc._.textrank
print(tr.elapsed_time)
```

```python
for combination in doc._.phrases:
    print(combination.text, combination.rank, combination.count)
```

```python
from bs4 import BeautifulSoup
from urllib.request import urlopen
```

```python
def get_only_text(url):
    page=urlopen(url)
    soup=BeautifulSoup(page)
    text = '\t'.join(map(lambda p: p.text, soup.find_all('p')))
    print(text)
    return soup.title.text,text
```

```python
url="https://en.wikipedia.org/wiki/Natural_language_processing"
text=get_only_text(url)
```

```python
from sumy.parsers.html import HtmlParser
from sumy.parsers.plaintext import PlaintextParser
from sumy.nlp.tokenizers import Tokenizer
from sumy.summarizers.lsa import LsaSummarizer
from sumy.nlp.stemmers import Stemmer
from sumy.utils import get_stop_words
from sumy.summarizers.luhn import LuhnSummarizer
```

```python
LANGUAGE="english"
SENTENCES_COUNT=10

url = "https://en.wikipedia.org/wiki/Natural_language_processing"
parser=HtmlParser.from_url(url,Tokenizer(LANGUAGE))
summarizer=LsaSummarizer()
summarizer=LsaSummarizer(Stemmer(LANGUAGE))
summarizer.stop_words=get_stop_words(LANGUAGE)
for sentence in summarizer(parser.document, SENTENCES_COUNT):
    print(sentence)
```

```python
text = """Knowledge about the structure and function of previous coronaviruses 
causing diseases like severe acute respiratory syndrome (SARS) and Middle East
respiratory syndrome (MERS) accelerated the development of various vaccine
platforms in early 2020.[1] In 2020, the first COVID19 vaccines were developed 
and made available to the public through emergency authorizations[2] and conditional 
approvals.[3][4] However, immunity from the vaccines was found to wane over time, 
requiring people to get booster doses of the vaccine to maintain protection 
against COVID19."""
```

```python
from sumy.parsers.plaintext import PlaintextParser
from sumy.nlp.tokenizers import Tokenizer
```

```python
parser=PlaintextParser.from_string(text,Tokenizer("english"))
```

```python
from sumy.summarizers.lex_rank import LexRankSummarizer
from sumy.utils import get_stop_words
summarizer_lex=LexRankSummarizer()

summarizer_lex.stop_words=get_stop_words("english")
summary=summarizer_lex(parser.document,5)
lex_summary=""
for sentence in summary:
    lex_summary += str(sentence)
print(lex_summary)
```
