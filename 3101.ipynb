```python
import spacy
nlp = spacy.load("en_core_web_sm")
```

```python
text = """Elon Musk is a visionary entrepreneur known for his groundbreaking 
    work in multiple industries. As the CEO of Tesla, SpaceX, and other ventures,
    he has pushed the boundaries of innovation. Musk is renowned for revolutionizing
    electric vehicles with Tesla, making space exploration more accessible through 
    SpaceX
"""
```

```python
doc = nlp(text)
for ent in doc.ents:
    print(f"Entity{ent.text},Label:{ent.label_}")
```

```python
from spacy import displacy
displacy.render(doc,style="ent")
```
